```python
import pandas as pd
```

```python
sampledf = pd.read_csv("https://raw.githubusercontent.com/vigneshk/Admission-Dataset/master/Admission.csv")
```

```python
import pymongo
```

```python
dbConn = pymongo.MongoClient("mongodb://localhost:27017/")
```

```python
dbname = "Ineuron"
```

```python
db=dbConn[dbname]
```

```python
collection_name='Admission_Details'

collection=db[collection_name]
```

```python
sampledf.reset_index(inplace=True)
```

```python
sampledf_dict = sampledf.to_dict("records")
```

```python
collection.insert_many(sampledf_dict)
```

```python
show_results = collection.find()
```

```python
for i in show_results:
    print(i)
```
